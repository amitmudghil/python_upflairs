```python
df = pd.read_csv(r'suicide_data/suicide_data.csv', index_col = 0)
df.head()
```

```python
df['State/UT'].str.count('a')#returns count of letter a
df['State/UT'].str.contains('a')
df['State/UT'].str.find('a') # find the position first occurence of the letter 
df['State/UT'].str.get_dummies() # create unique dataframe for all values of state/UT column
df['State/UT'].str.get(-1)
df['State/UT'].str.join('+')
df['State/UT'].str.lower()
df['State/UT'].str.count('a')
#df['State/UT'].str.repeat([3,4,2,4,2,4,2,2,5,7,3,3,5,2,3,5,4,5,6,7])
df['State/UT'].str.replace('a','k',case=False)
df['State/UT'].str.upper()
df['State/UT'].str.split().str.get(0)
```
